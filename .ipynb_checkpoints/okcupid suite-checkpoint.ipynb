Second step: create a model to find the best mate.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


data=pd.read_csv('okcupid_treat.csv')

# I have made a little mistake on essay variables. In fact it has to be changed
# like nominal and categorical variables.

data=pd.get_dummies(data=data, columns=['corpus'])

print(data.shape)
print(data.head(5))
print(data.columns)

# there are still nan values so we get rid of it before machine learning.

data=data.dropna(axis=0)
print(data.shape)
       
        
```

 Now with Kneighbors framework, in fact the method part to calculate the distance between two samples, we will use it to find the best mate for a sample.The function mate finder will find for one sample the first, second and third best choice.

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, MinMaxScaler

# To find our model we will drop sign data, because it s not because
# people have the same sign that it will match better and anyway, i
# don't know how it works with signs...I m not astrologist in fact !

for val in data.columns:
    
    if val.find('sign')!=-1:
        
        data=data.drop([val], axis=1)

      

# this function will select, i hope, the best mate.
    
          
    
def mate_finder(data, single):
    
    var_names=data.columns
    
    # we select potential mates groups depends on orientation and sex data.
    
    if single['orientation_gay']==1:
        
        if single['sex']==0:
            
            data=data[(data.sex==0) & (data.orientation_gay==1)]
            
        elif single['sex']==1:
            
            data=data[(data.sex==1) & (data.orientation_gay==1)]
        
    
    elif single['orientation_bisexual']==1:
        
        data=data[data.orientation_bisexual==1]
        
    elif single['orientation_straight']==1:
        
        if single['sex']==0:
            
            data=data[(data.sex==1) & (data.orientation_straight==1)]
            
        elif single['sex']==1:
            
            data=data[(data.sex==0) & (data.orientation_straight==1)]
       
    
    # we normalize our data to avoid unbalanced influences between variables.
    
    preprocess=MinMaxScaler()
    data=preprocess.fit_transform(data)
    single=preprocess.transform(np.array(single).reshape(1, -1))
    data=pd.DataFrame(data)
    single=pd.DataFrame(single)
    
    
    # we find the best mates with the beginning method of kneighbors framework.
    
    single=list(single.iloc[0,:])
    
    mate_index=[]
    mate_distances=[]
    
    for row in range(data.shape[0]):
        
        mate_index.append(row)
        mate=list(data.iloc[row,:])
        distances_square=[]
        
        for val in range(len(single)):
            
            diff_squared=(single[val]-mate[val])**2
            distances_square.append(diff_squared)
            
        mate_distances.append(sum(distances_square)**0.5)
    
    finder=pd.DataFrame({'mate_list': mate_index, 'distance': mate_distances})
    finder_sort=finder.sort_values(by=['distance'])
    print(finder.head(5))
    print(finder_sort.head(5))
    first_best_mate=data.iloc[finder_sort.iloc[0, 0],:]
    second_best_mate=data.iloc[finder_sort.iloc[1, 0],:]
    third_best_mate=data.iloc[finder_sort.iloc[2, 0], :]
    
    first_best_mate=preprocess.inverse_transform(np.array(first_best_mate).reshape(1, -1))
    first_best_mate=list(first_best_mate.transpose())
    first_best_mate=pd.DataFrame({'variables': var_names, 'data': first_best_mate})
       
    second_best_mate=preprocess.inverse_transform(np.array(second_best_mate).reshape(1, -1)) 
    second_best_mate=list(second_best_mate.transpose())
    second_best_mate=pd.DataFrame({'variables': var_names, 'data': second_best_mate})
    
    third_best_mate=preprocess.inverse_transform(np.array(third_best_mate).reshape(1, -1))
    third_best_mate=list(third_best_mate.transpose())
    third_best_mate=pd.DataFrame({'variables': var_names, 'data': third_best_mate})
        
    return first_best_mate, second_best_mate, third_best_mate
    

mate_finder(data, data.iloc[426])      
       
        
 
    
    
    
    
                
                
                
                
        
        
        

```
